```python
import matplotlib.pyplot as plt
labels=['harmless', 'type-unsupported', 'suspicious', 'confirmed-timeout', 'timeout', 'failure', 'malicious', 'undetected']
sizes=[0, 4, 0, 0, 0, 0, 0, 72]
```

```python
fig = plt.figure(figsize=(6, 6))
fig.patch.set_facecolor('white')
plt.pie(sizes, wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' })
# plt legend outside of plot right
plt.legend(labels=labels, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
